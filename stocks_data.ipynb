### Question 2 -  visualise the companies by segregating them according to Market Cap in three categories: Large Cap(>20,000 crore), Mid Cap (5,000 to 20,000 crore) & Small Cap(<5,000)

```python
above_20 = df[(df['Market Cap']>20000) | (df['Market Cap'] == 20000)].count()[3]

below_5 = df[(df['Market Cap']<5000) | (df['Market Cap'] == 5000)].count()[3]
below_5
```

```python
bw_5_20 = df[(df['Market Cap']<20000) &
            (df['Market Cap']>5000)].count()[3]

bw_5_20
```

```python
m_c = pd.DataFrame([['Above 20k',above_20],
                  ['Between 20 and 5',bw_5_20],
                   ['Below 5',below_5]], columns=['category', 'number_of_companies'])
m_c
```

```python
sns.barplot(data = m_c, x='category', y='number_of_companies')
plt.show()
```
